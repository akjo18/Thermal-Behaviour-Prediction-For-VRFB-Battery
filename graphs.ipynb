```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel("5a_40A.xlsx")
df2 = pd.read_excel("5a_45A.xlsx")
df3 = pd.read_excel("5a_50A.xlsx")
df4 = pd.read_excel("5a_60A.xlsx")
df5 = pd.read_csv("5b_40A.csv")
df6 = pd.read_csv("5b_45A.csv")
df7 = pd.read_csv("5b_50A.csv")
df8 = pd.read_csv("5b_60A.csv")
```

```python
x1 = df1.iloc[:,0].values[0::3]
y1 = df1.iloc[:,1].values[0::3]
x2 = df2.iloc[:,0].values[0::3]
y2 = df2.iloc[:,1].values[0::3]
x3 = df3.iloc[:,0].values[0::3]
y3 = df3.iloc[:,1].values[0::3]
x4 = df4.iloc[:,0].values[0::3]
y4 = df4.iloc[:,1].values[0::3]
x5 = df5.iloc[:,0].values[0::10]
y5 = df5.iloc[:,1].values[0::10]
x6 = df6.iloc[:,0].values[0::10]
y6 = df6.iloc[:,1].values[0::10]
x7 = df7.iloc[:,0].values[0::10]
y7 = df7.iloc[:,1].values[0::10]
x8 = df8.iloc[:,0].values[0::10]
y8 = df8.iloc[:,1].values[0::10]
print(len(x1))
print(len(x8))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(x1,y1,s=1,color='r',label="40A Charging (C1)")
ax1.scatter(x2,y2,s=1,color='b',label="45A Charging (C2)")
ax1.scatter(x3,y3,s=1,color='g',label="50A Charging (C3)")
ax1.scatter(x4,y4,s=1,color='y',label="60A Charging (C4)")


plt.xlabel("Time in hours")
plt.ylabel("VRFB Stack Temperature (Â°C)")
plt.text(3.5,23,'Flow rate = 10 L/min', fontsize = 10)
plt.text(3.3,44,'C4', fontsize = 10)
plt.text(3.8,39,'C3', fontsize = 10)
plt.text(4.5,35.5,'C2', fontsize = 10)
plt.text(5.25,30.5,'C1', fontsize = 10)

plt.grid()
plt.legend(loc=4)
plt.show()
```
